### An overview of statistical learning

Supervised tools: Involves building a statistical model for predicting, or estimating, an output based on one or more inputs.

Unsupervised tools: There are inputs but no supervising output, we can learn relationships and structure from such data

The Wage data involves predicting a *continuous* or *quantitative* output value. This is often referred to as a regression problem. However, in certain cases we may instead wish to predict a non-numerical value—that is, a *categorical* or *qualitative* output.

A classification problem in machine learning is a type of supervised learning task where the goal is to predict a discrete, categorical class label for a given input, based on training data (thanks Gemini lmao, its true, we want to predict something discrete basically, the discrete values are said 'Categories')

Finally, the last sort of problem we find are *clustering problems*. This is meant to group data points into "similar" categories. The example given is a group of customers (rows in a table), with different properties (variable values), we may want to group them into, well, groups.
