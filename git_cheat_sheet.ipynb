# Git & Terminal Cheat Sheet for Beginners

This notebook serves as a quick reference guide for navigating directories and managing version control using Git-Bash and GitHub.

---

## 📁 Part 1: Terminal Basics (Moving Around Your Computer)
Before running Git commands, you must tell the terminal which folder you want to look at.

*   **`pwd`** (Print Working Directory)
    *   *What it does:* Shows the exact folder path you are currently standing in.
*   **`ls`** (List)
    *   *What it does:* Lists all files and folders inside your current directory.
*   **`cd <path>`** (Change Directory)
    *   *What it does:* Moves you into a different folder.
    *   *Example:* `cd 'C:/Marcos stuff/ISLP/Stats mini projects/Titanic Kaggle'`
    *   *Pro Tip:* Type `cd ` (with a space) and drag-and-drop any folder from Windows File Explorer directly into the terminal window.
*   **`cd ..`**
    *   *What it does:* Moves you backward out of the current folder (up one level).
*   **Tab Autocomplete:** Type the first few letters of a folder name (e.g., `cd Titan`) and press the **Tab** key to automatically finish typing the name.

## 🛠️ Part 2: Repository Setup & Branching
These commands are used when establishing a connection between a local project folder and GitHub.

*   **`git init`**
    *   *What it does:* Turns an ordinary folder on your computer into a tracked Git repository. Run this if you get a `fatal: not a git repository` error.
*   **`git branch -M main`**
    *   *What it does:* Force-renames (`-M`) your active local branch from the older default name (`master`) to the modern GitHub default name (`main`) so they sync up cleanly.
*   **`git remote add origin <URL>`**
    *   *What it does:* Links your local repository to your remote GitHub repository using its unique `.git` URL.
*   **`git config --global user.name "Mescoman"`**
    *   *What it does:* Sets your global commit username display name.
*   **`git config --global user.email "your_email@example.com"`**
    *   *What it does:* Links your local machine's commits to your primary GitHub account profile.

## 🔄 Part 3: The Daily Workflow (Save & Upload)
This is the standard cycle you will run every time you modify your notebooks or scripts and want to back them up online. Remember: running `git push` by itself *only* uploads changes that have already been staged and committed!

| Command | Purpose | Analogy |
| :--- | :--- | :--- |
| **`git status`** | Displays modified, untracked, or staged files. | Checking your project shopping list. |
| **`git add .`** | Stages all current changes to be saved (the `.` means everything). | Packing your files into a shipping box. |
| **`git commit -m "Message"`** | Permanently snapshots staged changes with a short note. | Taping the box shut and labeling it. |
| **`git push`** | Uploads all localized commits safely to GitHub. | Loading the box onto the delivery truck. |

> **Your End-of-Day Terminal Routine:**
> ```bash
> git add .
> git commit -m "Finished cleaning the Age column and engineering Family features"
> git push
> ```

## ⬇️ Part 4: Pulling & Reverting Changes
Use these commands when you need to sync your local project folder with GitHub, or discard local changes to match what is saved online.

*   **`git pull origin main`**
    *   *What it does:* Fetches any new updates from your GitHub repository and merges them into your local files. *(Note: This will block you if you have unsaved local work that conflicts).*
*   **`git checkout HEAD -- <filename>`**
    *   *What it does:* Discards local, uncommitted changes for one specific file and reverts it back to exactly how it looks in your last GitHub save.
    *   *Example:* `git checkout HEAD -- Titanic_Notebook_EDA_cleaner.ipynb`
*   **`git reset --hard HEAD`**
    *   *What it does:* The ultimate local reset button. Wipes out **all** uncommitted test edits or mistakes across your entire folder, resetting everything to your last successful commit.

## 🛑 Part 5: Emergency & Modifier Tools
Useful commands when things get stuck or require minor tweaks.

*   **`Ctrl + C`**
    *   The ultimate terminal escape hatch. Forcibly kills any stuck, frozen, or hanging process (like an unexpected password loop) and returns your prompt.
*   **`git commit --amend --reset-author --no-edit`**
    *   Overwrites your immediate last commit with your updated profile username/email if you accidentally committed under a local work profile.
*   **`git log --oneline`**
    *   Prints a clean, chronological history list of your past project checkpoints.

### Understanding Command Flags
*   **Single Dash (`-`):** Shorthand shortcut for a single-letter modifier (e.g., `-M` for moving/renaming branches).
*   **Double Dash (`--`):** Fully spelled out word modifier flags (e.g., `--global` or `--no-edit`) used to make advanced options readable.
